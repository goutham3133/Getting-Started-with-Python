# Project 2
The folder constits a sceret message which present by image files. For understand the message you have to remove the number in their naming. Then open the folder you will find the answer.
Please extract the zip file in this folder to update the original naming file before start to find the solution.

What to do:

    List all the file name in the provided folder.
    Renaming all the naming file by removig the number in their naming.

```python
import os

def reNaming():
    # Get all the file name in folder prank
    file_list = os.listdir(r"./prank")

    # Get the current path
    savePath = os.getcwd()
    print ("Current working folder", savePath)

    # Let move to the path of folder containing the image
    os.chdir(r"./prank")

    # Change file name by removing the number
    for fileName in file_list:
        print("Old file name", fileName)
        print("New file name", fileName.translate(None, "0123456789"))
        os.rename(fileName, fileName.translate(None, "0123456789"))

    # Come back the previous path
    os.chdir(savePath)
    
reNaming()
```
